```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
data = pd.read_csv(r"C:\\Users\\Giorgis\\Desktop\\Projects\\ADTA coursework\\partI\\Census.csv")
data.head()
```

```python
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(20, 15))

# plt.figure(figsize=(18, 12))

sns.set_style('whitegrid')

sns.countplot(x='Sex', data=data, palette=["#2e5cff", "#ff692e"], ax=axes[0])

axes[0].set_ylabel('Total number of occurences', size=20, fontweight='bold')
axes[0].set_xlabel('Gender', size=20, fontweight='bold')
axes[0].set_title('Number of occurrences per Gender', size=30, fontweight='bold')
# Altering the x axis labels values' names
axes[0].set_xticklabels(('Male', 'Female')) 

plt.setp(axes[0].get_xticklabels(), size=18)
plt.setp(axes[0].get_yticklabels(), size=15)


sns.countplot(x='Occupation', data=data, hue='Sex', palette=["#2e5cff", "#ff692e"], ax=axes[1])


axes[1].set_xlabel('Occupations', size=20, fontweight='bold')
axes[1].set_ylabel('Number of occurencies', size=20, fontweight='bold')
axes[1].set_title('Number of gender occurrences\namong the occupations',
                  size=30, fontweight='bold')

# Altering the x axis labels values' names
axes[1].set_xticklabels(('No code required', 'Managers, Directors\nand Senior Officials', 'Professional',
                        'Associate Professional\nand Technical', 'Administrative\nand Secretarial',
                         'Skilled Trades', 'Caring, Leisure\nand Other Service', 'Sales and\nCustomer Service',
                        'Process, Plant and\nMachine Operatives', 'Elementary')) 

plt.setp(axes[1].get_xticklabels(), rotation=30, ha='right', size=18)
plt.setp(axes[1].get_yticklabels(), size=15)


plt.tight_layout()
sns.despine()

#Editing the legeng box's name values
leg = plt.legend(fontsize='xx-large')
regd = ['Male', 'Female']
i=0
for reg in regd:
    leg.get_texts()[i].set_text(reg)
    i+=1

    
fig.tight_layout()
sns.despine()

```
