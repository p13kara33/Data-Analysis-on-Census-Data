```python
regions = {'North East':'E12000001', 'North West': 'E12000002', 'London': 'E12000007',
           'West Midlands': 'E12000005', 'Yorkshire and The Humber': 'E12000003',
           'South West': 'E12000009', 'East Midlands': 'E12000004',  
           'South East':'E12000008', 'East of England': 'E12000006', 
           'Wales': 'W92000004'}
```

```python
data_region = data['Region']
df_region = pd.DataFrame(columns=['Region_names', 'Region'])
df_region['Region'] = data_region

for keys, values in regions.items():
    df_region.loc[df_region.Region == values, 'Region_names'] = keys

df_region = df_region.drop(columns='Region')
```

```python
# df = pd.merge(df_region, data, on='Region')
df = (df_region, data)
df = pd.concat(df, axis=1)
df


# df = df.to_frame().reset_index()
```

```python
no_of_hours_mean = round(np.mean(data['No of hours']),2)

data['No of hours'] = data['No of hours'].replace(np.nan, no_of_hours_mean)
```

```python
uk_regions = json.load(open('uk_regions_map.geojson', 'r'))
uk_regions
```

```python
uk_regions['features'][11]['properties']
```

```python
region_id_map = {}

for feature in uk_regions['features']:
    feature['id'] = feature['properties']['cartodb_id']
    region_id_map[feature['properties']['name']] = feature['id']

region_id_map.pop('Scotland')
region_id_map.pop('Northern Ireland')
region_id_map
```

```python
df['id'] = df['Region_names'].apply(lambda x: region_id_map[x])
df
```

```python
colorscale = ["#f7fbff", "#ebf3fb", "#deebf7", "#d2e3f3", "#c6dbef", "#b3d2e9", "#9ecae1",
    "#85bcdb", "#6baed6", "#57a0ce", "#4292c6", "#3082be", "#2171b5", "#1361a9",
    "#08519c", "#0b4083", "#08306b"]
```

```python
fig = px.choropleth_mapbox(df, 
                           locations='id', 
                           geojson=uk_regions, 
                           color='No of hours',
                           hover_name='Region_names',
                           hover_data=['No of hours'],
                           mapbox_style='carto-positron',  
                           color_continuous_scale="haline",
                           center={"lat": 53.5, "lon":-0.11},
                            
                           zoom=5, opacity=0.5)
fig.update_geos(fitbounds='locations', visible=False)

```
