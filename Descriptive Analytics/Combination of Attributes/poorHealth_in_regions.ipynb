```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
data = pd.read_csv(r"C:\\Users\\Giorgis\\Desktop\\Projects\\ADTA coursework\\partI\\Census.csv")
data.head()
```

# Poor health on regions

```python
poor_health = data[(data['Health'] >= 4)]
old_age = data[(data['Age'] >= 6)]
```

```python
plt.figure(figsize=(9, 6))

sns.set_style('whitegrid')

ax = sns.countplot(x='Health', data=poor_health, hue='Region', palette=["#ff692e", "#2e5cff",
            "#ffd12e", "#ff2e5c", "#5cff2e", "#ff0000", "black", "#8b4513",
            "#ff7f50", "#ffff80"])


ax.set_xlabel('Poor Health\'s values', size=20, fontweight='bold')
ax.set_ylabel('Number of occurencies', size=20, fontweight='bold')
ax.set_title('Number of occurencies of poor health \nper region', size=30, fontweight='bold')

# Altering the x axis labels values' names
ax.set_xticklabels(('Bad health', 'Very bad health')) 

plt.setp(ax.get_xticklabels(), size=11)

plt.tight_layout()
sns.despine()

#Editing the legeng box's name values
leg = plt.legend(fontsize='x-large')
regd = ['North East', 'North West', 'Yorkshire and the Humber', 'East Midlands', 'West Midlands', 'South West', 
        'East of England','South East', 'London', 'Wales']
i=0
for reg in regd:
    leg.get_texts()[i].set_text(reg)
    i+=1

plt.savefig('poorHealth_in_region.jpeg')
```

```python
plt.figure(figsize=(19, 12))

axes = sns.countplot(x='Age', data=old_age, hue='Region', palette=["#ff692e", "#2e5cff",
            "#ffd12e", "#ff2e5c", "#5cff2e", "#ff0000", "black", "#8b4513",
            "#ff7f50", "#ffff80"])
# plt.setp(ax.get_xticklabels(), rotation=70)
axes.set_xlabel('Age', size=20, fontweight='bold')
axes.set_ylabel('Total number', size=20, fontweight='bold')
axes.set_title('Age groups\' number of occurences in every Region', size=30, fontweight='bold')
#Altering the x axis labels values' names
axes.set_xticklabels(('55 to 64', '65 to 74', '75 and over' ))
plt.setp(axes.get_xticklabels(), size=20)
#Editing the legeng box's name values
leg = plt.legend(fontsize='x-large')
regd = ['North East', 'North West', 'Yorkshire and the Humber', 'East Midlands', 'West Midlands', 'South West', 
        'East of England','South East', 'London', 'Wales']
i=0
for reg in regd:
    leg.get_texts()[i].set_text(reg)
    i+=1

    
sns.despine()

plt.savefig('old_age_in_region.jpeg')
```
