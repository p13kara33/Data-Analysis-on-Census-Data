# Why is the mean of the age-group attribute lower than expected in London?

```python
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(25, 20))

# plt.figure(figsize=(18, 12))

sns.set_style('whitegrid')


sns.boxplot(x='Region', y='Age', data=data,  palette=["#ff692e", "#2e5cff",
            "#ffd12e", "#ff2e5c", "#5cff2e", "#ff0000", "black", "#8b4513",
            "#ff7f50", "#ffff80"], ax=axes[0])
axes[0].set_xlabel('Region', size=20, fontweight='bold')
axes[0].set_ylabel('Age', size=20, fontweight='bold')
axes[0].set_title('Boxplot of Age-groups among the Regions', size=30, fontweight='bold')

axes[0].set_yticklabels(('0 to 15', "16 to 24", '25 to 34', '35 to 44', '45 to 54', '55 to 64', '65 to 74', '75 and over' ))
plt.setp(axes[0].get_yticklabels(), size=15)
axes[0].set_xticklabels(('North East', 'North West', 'Yorkshire and the Humber', 'East Midlands', 'West Midlands', 'South West', 
        'East of England','South East', 'London', 'Wales')) 

plt.setp(axes[0].get_xticklabels(), size=17)


sns.countplot(x='Age', data=data, hue='Region', palette=["#ff692e", "#2e5cff",
            "#ffd12e", "#ff2e5c", "#5cff2e", "#ff0000", "black", "#8b4513",
            "#ff7f50", "#ffff80"], ax=axes[1])
# plt.setp(ax.get_xticklabels(), rotation=70)
axes[1].set_xlabel('Age', size=20, fontweight='bold')
axes[1].set_ylabel('Total number', size=20, fontweight='bold')
axes[1].set_title('Age groups\' number of occurences in every Region', size=30, fontweight='bold')
#Altering the x axis labels values' names
axes[1].set_xticklabels(('0 to 15', "16 to 24", '25 to 34', '35 to 44', '45 to 54', '55 to 64', '65 to 74', '75 and over' ))
plt.setp(axes[1].get_xticklabels(), size=20)
#Editing the legeng box's name values
leg = plt.legend(fontsize='x-large')
regd = ['North East', 'North West', 'Yorkshire and the Humber', 'East Midlands', 'West Midlands', 'South West', 
        'East of England','South East', 'London', 'Wales']
i=0
for reg in regd:
    leg.get_texts()[i].set_text(reg)
    i+=1

    
fig.tight_layout()
sns.despine()

plt.savefig('Age_in_Region.jpeg')
```

According to the box plot the mean of the age groups among every region apart from the one of London is the same. With the mean being the 45 to 54 age group and in London being the 35 to 44.

Taking into account the frequence table 1 and in combination with the third graph of the picture; which decribes how often every age group appears in every region, it is clear that the greater proportion of occurencies are emerging from the two most often Regions, as it is indisputably shown in the age groups of: 0 to 15, 16 to 24 and 45 to 54. However in the age group of 35 to 44 there is a minor anomally in the dataset as the London's bar is slightly higher than the one of the South East Region which has more occurencies in total according to the second graph. In addition the number that appears in the age group of 25 to 34 about London is substantially bigger than expected, as it exceeds the 16.000. Actually the 25 to 34 age group that is located in London consists the most often region-age goup variable combination. This contributes to a lower mean of the age group attribute in the Region of London.
