```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder

from mlxtend.frequent_patterns import fpgrowth
```

```python
data = pd.read_csv(r'C:\Users\Giorgis\Desktop\Projects\ADTA coursework\partI\Data_str.csv')
data = data.drop(['No of hours'], axis=1)
data.head(2)
```

```python
dataset = list((data.values))
len(dataset)

dataset
```

```python
te = TransactionEncoder()
te_ary = te.fit(dataset).transform(dataset)
df = pd.DataFrame(te_ary, columns=te.columns_)
df.head()
```

```python
support_of_50 = fpgrowth(df, min_support=0.1, use_colnames=True)
```

```python
support_of_50
```

### <font color=#225e71> Creating a dictionary of 2 or more correlated "descriptions" of a person that appear into in the dataset more oftenly </font>

```python
supp_dict = {}

for i in range(0, len(support_of_50)):
    if (len(support_of_50['itemsets'][i]) > 1):
        supp_dict[support_of_50['support'][i]] = support_of_50['itemsets'][i]
        
supp_dict
```
