```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

from apyori import apriori
```

```python
data = pd.read_csv(r'C:\Users\Giorgis\Desktop\Projects\ADTA coursework\partI\Data_str.csv')
data = data.drop('No of hours', axis=1)
data
```

```python
dataset = list((data.values))
len(dataset)

dataset
```

```python
rules = apriori(dataset, min_support = 0.5, min_confidence = 0.2, min_lift = 3, min_length = 2)
```

```python
results = list(rules)
results = pd.DataFrame(results)
results.head()
```
